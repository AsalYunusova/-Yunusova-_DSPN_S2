## The Moderating and Mediating Role of Mindfulness Mechanisms on Distress Tolerance 

### Introduction

   Recent research on distress tolerance has linked it to the development and maintenance of psychopathology such as substance abuse disorder, depression, and anxiety (Leyro, Zvolensky, Bernstein, 2010). Increasingly popular treatments for these disorders include acceptance and commitment therapy and mindfulness-based interventions (MBIs), which both promote acceptance skills. More specifically, mindfulness is "a way of paying attention to present moment experiences with an attitude of openness, acceptance, and equanimity"(Lindsay and Creswell, 2017). MBIs have been found to effectively increase distress tolerance among clinical populations (Cherkin et al., 2016; Morone et al., 2016). However, it remains unclear as to what specific mindfulness related mechanisms are driving these outcomes. A few studies have found that mindfulness increases people's distress tolerance during physical stress tasks such as the cold pressor task (participants are asked to submerge their hand in extremely cold water for as long as possible). Masedo and Esteve (2007), observed longer tolerance times and lower distress ratings among participants who received acceptance skills training gcompared to those who received suppresion and spontaneous coping training. Liu et al., (2013) found that mindfulness training was related to increased pain tolerance and distress ratings compared to the groups that received distraction and spontaneous coping intstructions. Carpenter, Sanford, and Hofmann (2019) also demonstrated that trait mindfulness was not a signifcant moderator between receiving mindfulness training and persisting in the cold pressor task. Additonally, they revealed significant indirect effects of mindfulness training on cold pressor task persistence. Lastly, Evans et al., (2014) attempted to dismantle mindfulness specific mechanisms that drive distress tolerance outcomes by creating two mindfulness groups: an observe-only group and an observe, describe and accept group but did not differences in pain tolerance between the two groups.  
   
   To build on previous research, we approach this gap in the literature with the theoretical framework of Monitor and Acceptance Theory (MAT) (Lindsay and Creswell, 2017). MAT posits that there are two key mechanisms that promote positive health benefits found in MBIs: monitoring ones experience and being accepting of that experience. Furthermore, MAT argues that the acceptance piece is a critical component. Indeed, many dismantling studies have shown that participants (stressed community adults, lonely adults), who are assigned to a mindfulness intervention focused on monitoring and acceptance skills (MA) show significantly greater stress and pain reductions compared to an intervention focused only on monitoring skills (where the acceptance piece is removed).
   
   The current study aims to investigate the relationship between trait distress tolerance, state mindfulness and equanimity (i.e., acceptance) on distress tolerance. We define distress tolerance as the "perceived capacity to withstand negative emotional and/or aversive states and the behavioral act of withstanding distressing internal states caused by some stressor" (Zvolensky, 2010). Participants (undergraduate CMU students), complete an online baseline questionnaire with psychological surveys including measures of trait distress tolerance, state mindfulness, and state equanimity. Participants then complete a baseline cold-pressor task where they submerge their hand into cold water for 5 seconds and provide a subjective distress rating upon removing their hand (see Figure 1). Then participants are randomly assigned to either a Monitor + Accept (MA) mindset condition (n=12) or a Monitor Only (MO) mindset condition (n = 13). The MA condition receives mindfulness training focused on monitoring unpleasant or uncomfortables thoughts and physical sensations that might arise during the upcoming cold-pressor task while the MO condition receives instructions to just focus and monitor the unplesanat or uncomfortable thoughts and sensations. Participants then are asked to submege their hand for as long as possible. They provide distress ratings every 30 seconds up to 5 minutes, when they remove their hand, and again every 30 seconds for 2 minutes (recovery period). The length of time that their hand is submerged is also recorded. Participants then complete a post cold pressor survey on their state mindfulness and equanimity. (Note: data collection is still ongoing, a G* Power analysis was conducted before data collection became that 46 participants are needed).

### Variables of Interest

**Subject ID:** (ID)
<br>
**Acceptance Skills:** (Condition, Condition.Dummy) - participants are randomly assigned to either the MO or MA intervention, MA is coded as 1 and MO is coded as 0 so that receiveing acceptance skills training = 1.

**Baseline Measures (before cold-pressor task and mindset manipulation):**
 - **Trait Distress Tolerance:** (DTS_avg) - This is measured using the validated self-report 'Distress Tolerance Scale' which has 15 items with response options ranging on a likert scale from 1-5 (Leyro, et al., 2011).
 - **Baseline State Equanimity:** (ES_avg) - This is measured using the validated self-report 'Equanimity-16 Scale', which has 16 items with response options ranging on a likert scale from 1-5 (Rogers, Shires, and Cayoun, 2021).
 - **Baseline State Mindfulness:** (SMS_avg) - This is measured using the validated self-report 'State Mindfulness Scale', which has 21 items with response options ranging on a likert scale from 1-5 (Ruimi, et al., 2019).
 - **Baseline Distress Rating:** (B_Distress) - This is measured using a self-report visual rating chart (Figure 1) with 1 response option ranging on a likert scale from 0-10 along with the question "How distressed are you by the pain in your hand right now?). Participants submerge their hand in the water for 5 seconds and then provide their distress rating.
 
**Cold Pressor Task Measures:**
 - **Submersion Time:** (totaltime) - A recorded time (seconds) of how long the participant kept their hand in the water during the actual cold pressor task (they're told to keep their hand in for as long as possible).
 - **Average Distress Rating:** (Distress_avg) - Participants provide a subjective distress rating every 30 seconds during the cold-pressor task. Then their ratings are averaged across a total of 10 possible ratings.
 
**Post Cold Pressor Task Measures:**
 - **Recovery Time:** (RecoveryTime) - This represents the time points during the recovery period after the cold-pressor task. After participants remove their hand from the water, they provide ratings every 30 seconds for 2 minutes.
 - **Recovery Distress:** (DistressRating) - This is an average rating of participants' distress during the recovery period. Participants' ratings are averaged across the 5 recovery periods.
 - **Post State Equanimity:** (ES_T2_avg) - This is measured using a modified 13 item version of the 'Equanimity-16 Scale', the likert responses are the same. This measures the participant's state equanimity by referring to their experiences in the past 15 minutes (e.g., during the cold pressor task).
 - **Post State Mindfulness:** (SMS_T2_avg) - This measures the participant's state mindfulness by referring to their experiences in the past 15 minutes (e.g., during the cold pressor task).

### 2. Moderation Analysis

```r
install.packages("jtools")
library(jtools)
#create a subset data with just the MA condition
MA <- d1[ which(d1$Condition.Dummy=='1'), ]
fitMod_MA <- lm(Distress_Avg ~ SMS_T2_Avg*DTS_avg,data=MA)
summary(fitMod_MA)
```

**Monitor + Accept Condition**: Looking at the model above, the interaction term (SMS_T2_Avg:DTS_avg) is not statistically significant (p=0.16), therefore there is no evidence of a moderation. In other words, the association between trait distress tolerance and average distress rating does not vary across different levels of post state mindfulness. Interestingly, we see no signficant relationship between state mindfulness and average distress nor do we see a significant relationship between trait distress tolerance and average distress.

Additionally, it is difficult to form an interpretation of the estimates in the model because a value of 0 does not appear to be in the range of possible values for the predictor variables (post state mindfulness and trait distress tolerance). This is because both variables are computed from a survey that uses likert ratings ranging from 1 to 5. 

Therefore, we will mean center the two predictor variables before examining the interaction in the analysis.

```r
#mean center predictors since we have no values of 0
MA_mod<-gscale(data=MA, vars=c("DTS_avg","SMS_T2_Avg"), center.only = TRUE)
str(MA_mod) #check that centering worked
fitMod_MA2 <- lm(Distress_Avg ~ SMS_T2_Avg*DTS_avg,data=MA_mod)
summary(fitMod_MA2)
```

Looking at the interaction term (SMS_T2_Avg:DTS_avg), we see that the p-value is still the same as what we had prior to centering. Because the variables are centered around their means, our interpretation of the estimates change. Now, a one unit increase in post state mindfulness, corresponds on average to an increase in average distress by 0.40. A one unit increase in trait distress tolerance, corresponds on average to an increase in average distress by 0.42. The predictive relationship between trait distress tolerance and average distress is 0.41 for a person scoring at the mean for state mindfulness.

```r
# create subset with MO condition and run moderation
MO <- d1[ which(d1$Condition.Dummy=='0'), ]
fitMod_MO <- lm(Distress_Avg ~ SMS_T2_Avg*DTS_avg,data=MO)
summary(fitMod_MO)
#mean center predictors since we have no values of 0
MO_mod<-gscale(data=MO, vars=c("DTS_avg","SMS_T2_Avg"), center.only = TRUE)
str(MO_mod)
fitMod_MO2 <- lm(Distress_Avg ~ SMS_T2_Avg*DTS_avg,data=MO_mod)
summary(fitMod_MO2)
```

**Monitor Only Condition**. Looking at the interaction term (SMS_T2_Avg:DTS_avg), we see that the p-value has not changed prior to centering. Similar to the MA condition, there is no significant moderation present nor do we see any relationships between our predictors (post state mindfulness and trait distress tolerance) with average distress. When taking a closer look at the new estimates, we see that a one unit increase in state mindfulness, corresponds on average to an increase in average distress by 0.26. A one unit increase in trait distress tolerance, corresponds on average to a decrease in average distress by 1.53. 

### Mediation Analysis:

```r
fitM <- lm(ES_T2_Avg ~ Condition.Dummy, data=d1) #Step 1: IV on M, Condition predicting submersion time
fitY <- lm(totaltime ~ ES_T2_Avg + Condition.Dummy, data=d1) #Step 2: IV and M on DV, Condition and state equanimity predicting totaltime
summary(fitM)
summary(fitY)
fitMed <- mediate(fitM, fitY, treat="Condition.Dummy", mediator="ES_T2_Avg")
summary(fitMed)
```

The first model is showing us that there is no significant relationship between the mediator (post state equanimity) and the IV (receiving acceptance skills). The second model is showing us that when you include post state equanimity in the model, acceptance skills is not a predictor of submersion time. However, we do see a significant relation between post state equanimity and submersion time (p<0.05). Looking at the last model, we see no strong mediating effect. The ACME or the mediating/indirect effect is not significant, nor is the ADE (direct effect). This makes sense as to why we see that the total effect (which combines the direct and indirect effects) and the Prop. Mediated (the proportion of indirect effects relative to direct effects) as both insignificant.

### Linear Mixed Effects Model

```r
#make a long format dataframe for each condition to run LME
LongDistress <- melt(Cold_All, id.vars=c("ID", "Condition.Dummy"),measure.vars=c("Peak_Distress","Post_30s_Distress","Post_1m_Distress","Post_1m30s_Distress","Post_2m_Distress"), variable.name="RecoveryTime", value.name="DistressRatings")
LongDistress$RecoveryTime<-as.numeric(LongDistress$RecoveryTime)
head(LongDistress)
Long_MA <- LongDistress[ which(LongDistress$Condition.Dummy=='1'), ]
Long_MO <- LongDistress[ which(LongDistress$Condition.Dummy=='0'), ]

# plot the data by recovery time and then grouped by subjects to examine fixed effects within each condition
ggplot(Long_MA, aes(RecoveryTime, DistressRatings, color = ID)) + geom_point()+
    geom_smooth(method="lm",se=FALSE)+
scale_x_continuous(breaks = c(1,2,3,4,5), labels = c("0 sec", "30 sec","60 sec","90 sec","120 sec"))

ggplot(Long_MO, aes(RecoveryTime, DistressRatings, color = ID)) + geom_point()+
    geom_smooth(method="lm",se=FALSE)+
scale_x_continuous(breaks = c(1,2,3,4,5), labels = c("0 sec", "30 sec","60 sec","90 sec","120 sec"))
```

By looking at the plots above, we see that individual subjects’ initial recovery rating at 0 seconds differ. Additionally, their distress ratings across the recovery periods differ as well, with some individuals having more steep declines in distress ratings while others have more constant distress ratings across time. 

```r
#examine simple linear fit for MA condition
fe.fit_MA = lm(DistressRatings~RecoveryTime, data=Long_MA)
summary(fe.fit_MA)
```

**Monitor+Accept Condition**: We see a significant effect of recovery time on recovery distress. For every unit increase in recovery time, we would expect a decrease of 1.10 in recovery distress. Next, we will examine subject effects as a random effect, we expect that the effect of recovery time on recovery distress will vary by subject. 

```r
#run linear mixed effects model for MA condition
me.fit1_MA = lmer(DistressRatings ~ RecoveryTime + (RecoveryTime | ID), data=Long_MA) 
me.fit2_MA = lmer(DistressRatings ~ RecoveryTime + (1 | ID), data=Long_MA) 
print("*******me.fit1_MA********")
summary(me.fit1_MA)
print("*******me.fit2_MA********")
summary(me.fit2_MA)
```

**Monitor+Accept Condition**: We see that recovery time on recovery distress in the fixed effects of me.fit1_MA, has decreased to -4.42 compared to -5.73 in the simple linear model. This tell us that we are getting a more conservative model fit on the fixed effects. Next, we will run a model comparison to interpret the value of including the random effects in our model since the random effects report doesn't provide us a t-value.

```r
# compare the models using the Akaike information criterion (AIC)
ic = AIC(fe.fit_MA, me.fit1_MA, me.fit2_MA)
ic
diff(ic$AIC)
```

We see that me.fit1_MA has the lowest AIC, thus it's the best model choice for the MA data. The difference in AIC between fe.fit_MA and the other two models is minimal, therefore the mixed-effects model accounts for marginally more variance in recovery distress than the simple linear model.

```r
#examine simple linear fit for MO condition
fe.fit_MO = lm(DistressRatings~RecoveryTime, data=Long_MO)
summary(fe.fit_MO)
```

**Monitor Only Condition**: For the MO condition, we see a significant effect of recovery time on recovery distress. For every unit increase in recovery time, we would expect a decrease of 1.00 in recovery distress. Next, we will examine subject effects as a random effect, we expect that the effect of recovery time on recovery distress will vary by subject.

```r
#run linear mixed effects model for MO condition
me.fit1_MO = lmer(DistressRatings ~ RecoveryTime + (RecoveryTime | ID), data=Long_MO) 
me.fit2_MO = lmer(DistressRatings ~ RecoveryTime + (1 | ID), data=Long_MO)
print("*******me.fit1_MO********")
summary(me.fit1_MO)
print("*******me.fit2_MO********")
summary(me.fit2_MO)
```

We see that recovery time on recovery distress in the fixed effects of me.fit1_MO, has decreased to -5.04 compared to -5.73 in the simple linear model, suggesting that we are getting a more conservative model fit on the fixed effects. 

```r
# compare the models using the Akaike information criterion (AIC)
ic = AIC(fe.fit_MO, me.fit1_MO, me.fit2_MO)
ic
diff(ic$AIC)
```

We see that me.fit1_MO has the lowest AIC, thus it's the best model choice for the MO data. The difference in AIC between fe.fit_MO and the other two models is minimal, therefore the mixed-effects model accounts for marginally more variance in recovery distress than the simple linear model.

### Conclusion

Because data collection is still on going, the conclusions that we draw from the three analyses above are very preliminary. Post state mindfulness did not moderate the relationship between trait distress tolerance and average distress during the cold-pressor task. If this finding holds, this would partially support previous research that could not demonstrate trait mindfulness as a signifcant moderator between receiving mindfulness training and persisting in the cold pressor task (Carpenter et al., 2019). If we are able to find a signficant moderation, it might suggest that mindfulness in the moment is more crucial than a person's general ability to be mindful in improving distress.

Next, we could not determine that post state equanimity mediated the relationship between receiving acceptance skills training and submersion time. We predict a partial mediation once data collection is complete. We predict a significant indirect relationship between post state equanimity and submersion time, along with a direct relationship between receiving acceptance skills training and submersion time. The latter would support the previous finding of a significant indirect effect of mindfulness training on cold pressor task persistence (Carpenter et al., 2019). 

Lastly, we found that a linear mixed effects model demonstrated that subjects signficantly explained the variance in the relationship between recovery time and recovery distress. This suggests that individual or with-in subject differences may be important to examine in distress tolerance outcomes rather than just between group differences (MA vs. MO). Overall, we see similar patterns in the models between the MA and MO conditions which makes sense since Evans et al., (2014) did not see any group differences (observe-only vs. obseve, describe and accept) in pain tolerance. However, because the models are currently underpowered, no final inferences can be made.
